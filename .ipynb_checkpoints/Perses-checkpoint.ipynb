```python
#PREPROCCESSING 

def TextPreprocessing(text):
    """Normalize Text
    """
    text = str(text)
    text = unicodedata.normalize("NFKD", text)  # Normalize
    text = '\n'.join(text.splitlines())  # Unicode break lines

    # Convert to upper
    text = text.upper()  # Convert to upper

    # Take care of breaklines & whitespaces combinations due to beautifulsoup parsing
    text = re.sub(r'[ ]+\n', '\n', text)
    text = re.sub(r'\n[ ]+', '\n', text)
    text = re.sub(r'\n+', '\n', text)

    # To find MDA section, reformat item headers
    text = text.replace('\n.\n', '.\n')  # Move Period to beginning

    text = text.replace('\nI\nTEM', '\nITEM')
    text = text.replace('\nITEM\n', '\nITEM ')
    text = text.replace('\nITEM  ', '\nITEM ')

    text = text.replace(':\n', '.\n')

    # Math symbols for clearer looks
    text = text.replace('$\n', '$')
    text = text.replace('\n%', '%')

    # Reformat
    text = text.replace('\n', '\n\n')  # Reformat by additional breakline

    text = text.strip()  #<=== remove white space at the beginning and end
    text = text.replace('\n', ' ') #<===replace the \n (new line) character with space
    text = text.replace('\r', '') #<===replace the \r (carriage returns -if you're on windows) with space
    text = text.replace('&nbsp;', ' ') #<===replace "&nbsp;" (a special character for space in HTML) with space. 
    text = text.replace('&#160;', '') #<===replace "&#160;" (a special character for space in HTML) with space.
   
    #Pre-processing the html content by removing extra white space and combining then into one line.
    while '  ' in text:
        text = text.replace('  ', ' ') #<===remove extra space
    return text

   
```

```python
# READ HTML
txtSubPath = "./txt/" #<===The subfolder with the extracted text files
htmlSubPath = "./EDGAR/"
DownloadLogFile = "10kDownloadLog.csv" #a csv file (output of the 3DownloadHTML.py script) with the download history of 10-K forms
ReadLogFile = "10kReadlog.csv" #a csv file (output of the current script) showing whether item 1 is successfully extracted from 10-K forms

def readHTML(file_name):
    input_path = htmlSubPath+file_name
    output_path = txtSubPath+file_name.replace(".htm",".txt")
    input_file = open(input_path,'r+b')
    page = input_file.read()  #<===Read the HTML file into Python
    page=TextPreprocessing(page)
    #Using regular expression to extract texts that match a pattern
    #Define pattern for regular expression.
        #The following patterns find ITEM 1 and ITEM 1A as diplayed as subtitles
        #(.+?) represents everything between the two subtitles
    #If you want to extract something else, here is what you should change
    #Define a list of potential patterns to find ITEM 1 and ITEM 1A as subtitles   
    regexs = ('bold;\">\s*item 1\.(.+?)bold;\">\s*item 1A\.',   #<===pattern 1: with an attribute bold before the item subtitle
              'b>\s*Item 1\.(.+?)b>\s*Item 1A\.',               #<===pattern 2: with a tag <b> before the item subtitle
              'Item 1\.\s*<\/b>(.+?)Item 1A\.\s*<\/b>',         #<===pattern 3: with a tag <\b> after the item subtitle          
              'Item 1\.\s*Business\.\s*<\/b(.+?)Item 1A\.\s*Risk Factors\.\s*<\/b') #<===pattern 4: with a tag <\b> after the item+description subtitle 
    #Now we try to see if a match can be found...
    # Write the regex

    print("/n ===================NEXT===================/n")
    for regex in regexs:
        match = re.search(regex, page, flags=re.IGNORECASE)  #<===search for the pattern in HTML using re.search from the re package. Ignore cases.
        #If a match exist....
        print(match)
        if match:
            #Now we have the extracted content still in an HTML format
            #We now turn it into a beautiful soup object
            #so that we can remove the html tags and only keep the texts
            
            soup = BeautifulSoup(match.group(1), "html.parser") #<=== match.group(1) returns the texts inside the parentheses (.*?) 

            #soup.text removes the html tags and only keep the texts
            rawText = soup.text
            print(soup)
            #remove space at the beginning and end and the subtitle "business" at the beginning
            #^ matches the beginning of the text
            outText = re.sub("^business\s*","",rawText.strip(),flags=re.IGNORECASE)
            output_file = open(output_path, "w")
            output_file.write(outText)  
            output_file.close()            
            break  #<=== if a match is found, we break the for loop. Otherwise the for loop continues

    input_file.close()    
    return match
```

```python
def get_documents(text):
    """
    Extract the documents from the text

    Parameters
    ----------
    text : str
        The text with the document strings inside

    Returns
    -------
    extracted_docs : list of str
        The document strings found in `text`
    """

    final_docs = []
    start_regex = re.compile(r'<DOCUMENT>')
    end_regex = re.compile(r'</DOCUMENT>')
    
    start_idx = [x.end() for x in re.finditer(start_regex, text)]
    end_idx = [x.start() for x in re.finditer(end_regex, text)]
    
    for start_i, end_i in zip(start_idx, end_idx):
        final_docs.append(text[start_i:end_i])
    
    
    return final_docs
```

```python
def find_mda_from_text(text, start=0):
    """Find MDA section from normalized text
    Args:
        text (str)s
    """
    debug = False

    mda = ""
    end = 0

    # Define start & end signal for parsing
    item7_begins = [
        '\nITEM 7.', '\nITEM 7 –', '\nITEM 7:', '\nITEM 7 ', '\nITEM 7\n'
    ]
    item7_ends = ['\nITEM 7A']
    if start != 0:
        item7_ends.append('\nITEM 7')  # Case: ITEM 7A does not exist
    item8_begins = ['\nITEM 8']
    """
    Parsing code section
    """
    text = text[start:]

    # Get begin
    for item7 in item7_begins:
        begin = text.find(item7)
        if debug:
            print(item7, begin)
        if begin != -1:
            break

    if begin != -1:  # Begin found
        for item7A in item7_ends:
            end = text.find(item7A, begin + 1)
            if debug:
                print(item7A, end)
            if end != -1:
                break

        if end == -1:  # ITEM 7A does not exist
            for item8 in item8_begins:
                end = text.find(item8, begin + 1)
                if debug:
                    print(item8, end)
                if end != -1:
                    break

        # Get MDA
        if end > begin:
            mda = text[begin:end].strip()
        else:
            end = 0

    return mda, end
```

```python
def get_prices(name,start_date,end_date,timeframe):
    prices = []
    #Download selected Ticker data
    yahoo_financials = YahooFinancials(name)
    historical_stock_prices1 = yahoo_financials.get_historical_price_data(start_date,
    end_date, 
    timeframe)            
    prices = historical_stock_prices1[name]['prices']
    prices = pd.DataFrame(prices)
    #changing formatted_date object to datetime formate
    prices['formatted_date']=pd.to_datetime(prices['formatted_date'], format=('%Y-%m-%d'))
    prices = prices.drop('date', axis=1)
    #Capitalizing 1st letter of each column
    prices.columns = [string.capwords(i) for i in prices.columns ]
    prices = prices.rename(columns = {'Formatted_date':'Date','Adjclose':'Adjusted_close'}).set_index('Date')
    prices.insert(6, 'Returns', prices['Adjusted_close'].pct_change(1))
    return prices

def get_info(name):
    ticker = yf.Ticker(name) 
    info = Info()
    info.industry = ticker.info['industry']
    info.exchange = ticker.info['exchange']
    info.sector = ticker.info['sector']
    return info
```

```python
def downloadStockInfo(stocks,start_date,end_date,timeframe):
    inputs = Inputs()
    for stock in stocks:
        prices = get_prices(stock, start_date=start_date, end_date=end_date,timeframe=timeframe)
        info = get_info(stock)
        asset = Stock(name=stock, dailyPrices=prices)
        asset.info = info
        inputs.assets.append(asset)
    return inputs

def downloadInputsTrends(inputs):
    for asset in inputs.assets:
        trends = Trends()
        args = []
        args.append(asset.name)
        pytrends.build_payload(args, cat=0, timeframe='today 12-m') 
        interest_over_time_df = pytrends.interest_over_time()
        interest_over_time_df.drop('isPartial', inplace=True, axis=1)
        trends.IOT = interest_over_time_df
        #Calculate standardized interest_over_time
        trends.mean_IOT = interest_over_time_df.mean()
        trends.std_IOT = interest_over_time_df.std()
        trends.IOT_standardized = (interest_over_time_df - trends.mean_IOT) / trends.std_IOT
        asset.trends = trends
    return inputs
```
